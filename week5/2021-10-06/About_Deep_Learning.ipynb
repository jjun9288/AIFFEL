# Machine learning vs Deep learning



### Deep learning is inspired by neural networks of the brain to build learning machines which discover rich and useful internal representations, computed as a composition of learned features and functions.

- 뇌의 신경 구조로부터 영감을 받아 신경망 형태로 설계된 딥러닝의 목표는, '합성된 함수를 학습시켜서 풍부하면서도 유용한 '내재적 표현'을 찾아내는 machine을 구축하는 것'이다.   
   
   
    - 내재적 표현? : 그 데이터가 담고 있는 정보의 총체, 혹은 함의를 나타내는 표현. 다만, raw한 sensory 데이터는 그 자체로는 중요한 의미지를 담지 못하고 있을 수 있다. 마치 이미지를 멀리서 보면 강아지가 보이지만, 각 픽셀 하나하나의 값들은 강아지와 연관이 없는 것처럼...

### 신경망의 본질


   
   


이는 결국 함수와 같다.   




1. 함수의 첫 번 째 역할 : Relation.
 - y = f(x) 에서 y는 x의 변화에 종속적이다. 즉 x가 변하면 그 변화하는 정도에 따라 y가 변하게 된다.   
   
   
2. 함수의 두 번 째 역할 : Transformation.   
   
   
3. 함수의 세 번 째 역할 : Mapping.   
    -One to one : 맥북 구매한 연에 따라 중고가를 출력하는 함수   
    -Many to one : 맥북의 CPU 사양, RAM 용량, 화면 크기 등을 입력 받아 중고가를 출력하는 함수    
    -Many to many : 128x128 이미지를 입력하면 각 카테고리에 대한 확률 값을 담은 벡터를 출력하는 함수

### 함수와 모델

1. 먼저 모델을 어떤 함수로 나타낼 것인지를 정해야 한다.
    - 입력 데이터가 하나인 경우 (One to one), f(x) = wx + b 와 같은 일차 함수 사용
    - 입력 데이터가 여러 개의 특징으로 이루어진 벡터인 경우 (Many to one), f(x) = wx(1) + wx(2) + ... + wx(n) 과 같은 다변수 선형 함수 사용
    - 신경망 모델 사용
    
2. 모델의 함수 형태를 정했다면, 해당 함수 공간 안에서 최적의 함수를 찾아나간다.
    - 이를 모델 학습이라 한다.대표적으로 딥러닝의 신경망 모델은 Gradient descent 를 사용한다.

### ML vs DL?

- 머신러닝은 주어진 데이터로부터 패턴을 학습해서 원하는 값을 예측하는 것에 방점이 있다면, 딥러닝은 입력된 데이터에 내재되어 있는 표현 그 자체를 나타내도록 학습하는 것이 목표이다.



### Is Deep Learning a Black Box?

- 블랙박스가 아니다.   
   
    - 신경망은 여러 개의 함수를 엮어 놓은 것이라 할 수 있고, 그 함수는 본질적으로 회귀 식과 형태가 같다.
    - 즉, 여러 함수의 연산은 간단한 계산으로 이루어져 있고, 어떤 원리로 동작하는지 알기 때문에 블랙박스라고 할 수 없다.
    
- 블랙박스가 맞다.   
   
    - 딥러닝의 구조는 매우 복잡하고 연산량 또한 엄청나므로 계산 과정을 전체적으로 이해하는 것은 불가능하다.
    - 학습된 딥러닝 모델이 어떤 결과를 내놓을 지 예측할 수 없고, 어떤 근거에 따라 판단되었는지 또한 설명할 수 없다.

### XAI (eXplainable AI)

- 머신러닝 모델에 대해 해석을 할 수 있어야 하는 이유
    - 모델이 예측한 결과가 어떠한 과정으로 도출되었는지 설명할 수 있다면 납득할만한 과정을 거쳐 도출되었다는 것을 확인함으로써 그 결과를 신뢰할 수 있다.
