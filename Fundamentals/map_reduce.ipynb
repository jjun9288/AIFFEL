### map

```python
my_num = ['1', '2', '3', '4']
my_num_int = list(map(int, my_num))    #my_num의 각 원소에 int 함수 적용
print(my_num_int)

my_num_square = list(map(lambda x : x*x, my_num_int))
print(my_num_square)
```

### filter

```python
my_num = range(-5, 5)
my_num_plus = list(filter(lambda x : x>0, my_num))
print(my_num_plus)
```

### reduce

- 컬렉션을 축약 (시그마)

```python
from functools import reduce
mynum = [1, 2, 3, 4, 5]
add = reduce((lambda x, y : x + y), mynum)

print(add)
```

### map reduce의 4단계 : split - map - shuffle - reduce





#### ex) 알파벳 수 찾기



1. 입력 데이터 - 'hello python'
2. split - h,e,l,l,o,( ),p,y,t,h,o,n
3. map - ('h',1), ('e',1), ('l',1), ('l',1), ('o',1), (' ',1), ('p',1), ('y',1), ('t',1), ('h',1), ('o',1), ('n',1)
4. reduce - {'h':2, 'e':1, 'l':2, 'o':2, ' ':1, 'p':1, 'y':1, 't':1, 'n':1}

```python
def mapper(text):
    split = []
    for i in text:
        split.append((i, 1))
    return split

text = 'hello python'

map = mapper(text)
print(map)
```

```python
def reducer(split):
    out = {}
    for i in split:
        if i[0] not in out.keys():
            out[i[0]] = 1
        else:
            out[i[0]] += 1
    return out
```

```python
reducer(mapper(text))
```
