```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a =  np.array([0.1, 0.2, 0.7])

def softmax(a):
    exp_a = np.exp(a)
    exp_total = np.sum(exp_a)
    result = exp_a / exp_total
    return result

print(softmax(a))
```

```python
x = np.linspace(-1,1,100)
y = softmax(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
