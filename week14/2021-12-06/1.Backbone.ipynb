# ResNet

### Degradation problem

딥러닝 모델의 레이어가 깊어졌을 때 모델이 수렴했음에도 불구하고 오히려 레이어 갯수가 적을 때보다 모델의 training/test error가 더 커지는 현상이 발생하는데, 이것은 overfitting 문제가 아니라 네트워크 구조상 레이어를 깊이 쌓았을 때 최적화가 잘 안되기 때문에 발생하는 문제이다.

### Residual Block

ResNet은 깊은 네트워크의 학습이 어려운 점을 해결하기 위해서 레이어의 입력값을 활용하여 레이어가 residual function을 학습하도록 한다.   
일종의 지름길을 통해서 레이어가 입력값을 직접 참조하도록 레이러를 변경했다고 볼 수 있다.





실제로 shortcut이 적용된 오른쪽에서 레이어가 깊어져도 학습이 잘 되는 효과를 볼 수 있다.

# DenseNet



- DenseNet은 ResNet의 shortcut connection을 마치 Fully Connected Layer처럼 촘촘히 가지도록 한다면 더욱 성능 개선 효과가 클 것이라고 생각한 것이다.   


- 일반적인 컨볼루션 네트워크가 L개의 레이어에 대해 각 레이어 간 하나씩의 연결을 하는 것과 달리 DenseNet은 L(L+1)/2개의 직접적인 연결을 만든다.   


- Shortcut connection이 있어 ResNet과 비슷해 보일 수 있지만 ResNet은 shortcut을 원소별로 단순히 더해주었던 반면, DenseNet은 하나하나를 차원으로 쌓아서 하나의 텐서로 만들어낸다는 사실이 다르다.   


- ResNet과 또 다른 점은 ResNet은 connection에 다른 연산이 없었지만 DenseNet은 이 텐서에 대해 배치 정규화, ReLU 활성화 함수, 3x3 컨볼루션 레이어를 통해서 pre-activation을 수행한다.

# SENet

SENet은 Squeeze and Excitation Networks의 줄임말이다.



- Squeeze를 통해 1x1xC 의 벡터를 얻는다. 이 벡터는 채널별 정보를 압축해 담고 있다.

- Excitate를 통해 채널별 정보에 따라 어떤 채널을 강조할지 판단한다.
